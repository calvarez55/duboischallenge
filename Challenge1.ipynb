```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("D:/documents copy/research/Practice/dubois challenge/challenge1.csv")
```

```python
df.head()
```

```python
fig, ax = plt.subplots(facecolor='blanchedalmond')

df.plot(x='Colored', y='Year', ax=ax, grid='true',
        color=('black'),
        xticks=(100,95,90,85,80,75,70,65,60,55,50,45,40,
               35,30,25,20,15,10,5,0),
        yticks=(1790,1800,1810,1820,1830,1840,1850,1860,1870,1880,1890),
       figsize=(8,12))
df.plot(x='White', y='Year',color=('black'),ls=('dashed'),ax=ax,grid='true')
ax.set_title('\nCOMPARATIVE INCREASE OF WHITE AND COLORED\n POPULATION OF GEORGIA.\n',
             fontsize=20, fontfamily='monospace')
ax.set_xlabel("\nPERCENTS")
ax.grid(color='lightpink')
ax.invert_xaxis()
ax.legend(["= COLORED", "= WHITE"], loc='lower center', 
          bbox_to_anchor=(.5,-.10),frameon=False, ncol=2)
ax.set_facecolor("blanchedalmond")
plt.text(53,1786,'}',rotation=270,fontsize=20)
plt.margins(x=0,y=0)

fig.savefig('D:/documents copy/research/Practice/dubois challenge/challenge1.png')
```
